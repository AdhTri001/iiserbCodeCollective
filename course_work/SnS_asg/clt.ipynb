```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
POP = np.random.random
S = 10000
N = 30

samples = POP((S, N))
```

```python
plt.hist(samples[0], bins=np.arange(0, 1.1, .1), alpha=.5, color='b', edgecolor='black')
plt.xticks(np.arange(0, 1.1, .1))
plt.yticks(np.arange(0, 6, 1))
plt.show()
```

```python
means = np.mean(samples, axis=1)
f = means - .5
b = np.arange(-.5, .6, .025)
plt.hist(f, bins=b, alpha=.5, color='b', edgecolor='black')
plt.xticks(b[::4])
plt.show()
```

```python
possibleVal = 0

for T in range(1000):

    tosses = np.random.randint(0, 2, (100000))

    arr = tosses == 1
    headsIndex = [i for i in range(100000) if arr[i]]

    avg = [(headsIndex[i + 100] - headsIndex[i]) for i in range(0, len(headsIndex) - 100)]

    count = 0

    for a in avg:
        if a >= 226:
            count += 1

    possibleVal += count / len(avg)

print(possibleVal / 1000)
```
