```python
import pandas as pd
import requests
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

R1 = requests.get("https://www.hackerrank.com/rest/contests/ignyte/leaderboard?limit=500", headers=headers).json()
R2 = requests.get("https://www.hackerrank.com/rest/contests/ignyte-r2/leaderboard?limit=500", headers=headers).json()
```

```python
df1 = pd.DataFrame(R1['models'])[['hacker', 'score']]
df2 = pd.DataFrame(R2['models'])[['hacker', 'score']]

df1['pointsR1'] = df1['score'] / 8000
df2['pointsR2'] = df2['score'] / 2300

df = pd.merge(df1, df2, on='hacker', how='outer')
df.fillna(0, inplace=True)

df['total'] = (df['pointsR1'] * .4 + df['pointsR2'] * .6) * 100
df.drop(columns=['pointsR1', 'pointsR2'], inplace=True)
df.sort_values(by='total', ascending=False, inplace=True)
df.set_index('hacker', inplace=True)

df
```
